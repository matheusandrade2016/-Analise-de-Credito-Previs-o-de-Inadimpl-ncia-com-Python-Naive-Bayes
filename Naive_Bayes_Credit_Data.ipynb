```python
# Carregando minha base de credit do downlod feito na seção 03
# importando a bibilioteca

# Criando minhas variaveis de treino e teste pela base de dados do credit.pkl
# rb -- > read

import pickle
with open('credit.pkl', 'rb') as f:
  x_credit_treinamento, y_credit_treinamento, x_credit_teste, y_credit_teste = pickle.load(f)
```

```python
# Importando as Bibiotecas
import pandas as pd
import numpy as np

# O algoritmo de Naive Bayes consiste em encontrar uma probabilidade a posteriori 
# (possuir a doença, dado que recebeu um resultado positivo), multiplicando a probabilidade ]
# a priori (possuir a doença) pela probabilidade de “receber um resultado positivo,
#  dado que tem a doença”.

from sklearn.naive_bayes import GaussianNB
```

```python
# Verificando a o shape da base treinamento
# 1500 -- > valores para treinamento na base X, sendo 3 colunas elas são (idade,renda e divida)
# 1500 -- > valores das classes Y sendo se a pessoa paga ou nao paga o emprestimo

x_credit_treinamento.shape, y_credit_treinamento.shape
```

```python
# Verificando o shape do meu teste 
# 500, 3 X == > previsores
# (500,)) Y ==> Classe

x_credit_teste.shape, y_credit_teste.shape
```

```python
# Criando meu codigo Naive Bayes para a gerção da tabela para a verificação se a pessoa ira pagar
# ou não ira pagar

naive_credit_data = GaussianNB()
naive_credit_data.fit(x_credit_treinamento, y_credit_treinamento)
```

```python
# Criando minha variavel Previsora sendo os 1500 registros que foram treinados e os 500 que nao 
# foram treinados
previsoes = naive_credit_data.predict(x_credit_teste)
```

```python
# Verificando minha variavel previsores

# Sendo assim ele classificou como 1 pessoas que não pagam
# Sendo assim ele classificou como 0 pessoas que pagam

previsoes
```

```python
# Agora comparando com os valors com minha base de dados real 
y_credit_teste
```

```python
# Importando minha biblioteca Skleanr
# from sklearn.metrics import accuracy_score - > ira contar quantos erros e acertos na base 
from sklearn.metrics import accuracy_score
```

```python
# Sendo dos 500 algoritimos ele acertou 93 % 

accuracy_score(y_credit_teste,previsoes)
```
